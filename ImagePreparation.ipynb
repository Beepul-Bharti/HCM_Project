```python
#Import the relevant libraries for importing images

# check Pillow version number
import PIL
print('Pillow Version:', PIL.__version__)
# load all images in a directory
from os import listdir
import glob
from matplotlib import image
import os
from matplotlib import pyplot as plt
```

```python
# load all images in a directory
loaded_images = list()
TopDirectory = listdir('Cropped_4ChamberCine')

for filename in TopDirectory:
    print(filename)
    for i in filename:
        if i == '0':
            continue
        else:
            path_it = 'C:\\Users\\Ryan\\Documents\\Sevde\\HCMData\\Cropped_4ChamberCine' + '\\' + filename + '\\' + i + '.png'
            img_data = image.imread(path_it)
            # store loaded image
            loaded_images.append(img_data)
```
